```python
def filter_nodes(input_filename, output_filename):
    # 读取文件
    with open(input_filename, 'r') as file:
        lines = file.readlines()

    # 过滤节点数小于3的行
    filtered_lines = [line.strip() for line in lines if len(line.split()) > 3]

    # 写入输出文件
    with open(output_filename, 'w') as output_file:
        for line in filtered_lines:
            output_file.write(line + '\n')

if __name__ == "__main__":
    input_filename = 'D:\Ace\PaperCode\DGE\growhon-master\clickstream.txt'
    output_filename = 'D:\Ace\PaperCode\DGE\growhon-master\clickstream_processed3.txt'
    filter_nodes(input_filename, output_filename)
    print(f"处理完成。输出文件名为{output_filename}")
```

```python
def count_unique_nodes(filename):
    unique_nodes = set()

    # 读取文件
    with open(filename, 'r') as file:
        lines = file.readlines()

    # 提取节点并添加到集合中
    for line in lines:
        nodes = line.strip().split()
        unique_nodes.update(nodes)

    # 统计不同节点数
    count = len(unique_nodes)

    return count

if __name__ == "__main__":
    input_filename = 'D:\Ace\PaperCode\DGE\growhon-master\clickstream_processed3.txt'
    unique_node_count = count_unique_nodes(input_filename)
    print(f"文件中不同的节点数有：{unique_node_count}个")
```
